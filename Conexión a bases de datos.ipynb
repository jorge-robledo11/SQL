```python
import pyodbc
import pandas as pd
import numpy as np
```

```python
# Información de la base de datos
server = 'ROBLEDO'
db = 'Maraton'
usuario = 'lynn'
password = '1234'

# Nos conectamos a la base de datos con SQL Server
try:
    conn = pyodbc.connect(driver='{SQL server}', host=server, database=db)
    print('Conexión exitosa')
except:
    print('Error de conexión')

# try:
#     conn = pyodbc.connect('DRIVER={OBDC DRIVER 17 for SQL server}; SERVER='+server+';DATABASE='+db+';UID='+usuario+';PWD='+password)
#     print('Conexión exitosa')
# except:
#     print('Error de conexión')
```

```python
# Filtramos la información utilizando SQL
maraton = pd.read_sql(
    """
    SELECT *
    FROM Maraton_NY
    WHERE age > 40 AND gender = 'Female' AND time < 250
    """,
    conn
)

conn.close()
```

```python
# Dataframe
maraton
```

```python
# Media
print(maraton.mean(numeric_only=True), '\n')

# Mediana
print(maraton.median(numeric_only=True), '\n')

# Varianza
print(maraton.var(numeric_only=True), '\n')

# Desviación estándar
print(maraton.std(numeric_only=True))
```

```python
# Transformando a datetime la variable 'time'
maraton['time'] = maraton['time'].apply(lambda x: pd.to_datetime(x))
```

```python
# Otra forma de capturar variables

# Capturar los tipos de variables
numericals = list(maraton.select_dtypes(include = ['int', 'float']).columns)
categoricals = list(maraton.select_dtypes(include = ['object', 'bool']).columns)
temporary = list(maraton.select_dtypes(include = ['datetime', 'timedelta']).columns)
discretes = [col for col in maraton[numericals] if len(maraton[numericals][col].unique()) < 20]
continuous = [col for col in maraton[numericals] if col not in discretes]

# Variables
print(f'Hay {len(numericals)} variables numéricas')
print(f'Hay {len(continuous)} variables continuas')
print(f'Hay {len(discretes)} variables discretas')
print(f'Hay {len(temporary)} variables temporales')
print(f'Hay {len(categoricals)} variables categóricas')
```
